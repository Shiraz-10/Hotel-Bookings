# Data Visualizations

```python
hotel_types = data["hotel"].value_counts()
```

```python
print(hotel_types)
```

```python
hotel_types.plot(kind='bar')
```

```python
# Group the data by month and count the number of bookings in each month
monthly_bookings = data['arrival_date_month'].value_counts().reindex(
    ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December'])

# Plotting the seasonality of bookings
plt.figure(figsize=(12, 6))
monthly_bookings.plot(kind='line', marker='o', color='purple')
plt.title('Seasonality of Bookings')
plt.xlabel('Month')
plt.ylabel('Number of Bookings')
plt.grid
```

```python
# Count the number of bookings by country
country_booking_counts = data['country'].value_counts()

# Plot the top 15 countries with the most bookings
plt.figure(figsize=(14, 7))
country_booking_counts.head(15).plot(kind='bar', color='black')
plt.title('Top 15 Countries by Number of Bookings')
plt.xlabel('Country')
plt.ylabel('Number of Bookings')
plt.xticks(rotation=45)
plt.show()
```

```python
# Plot the top 5 countries with the most bookings as a pie chart
plt.figure(figsize=(6, 6))
country_booking_counts.head(5).plot(kind='pie', autopct='%1.1f%%', colors=plt.cm.Paired.colors)
plt.title('Top 5 Countries by Number of Bookings')
plt.ylabel('')  # Remove the y-label for aesthetics
plt.show()
```
